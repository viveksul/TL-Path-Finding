```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Accuracy of N=30, with scaling parameter 1000 , the TL models also have same scaling parameter
Acc30 = 0.67
```

```python
# Accuracy of induvidually trained models for RBI 0.4, 0.5, 0.7 and 0.88 mean
Acc = [0.7000, 0.6700, 0.6430, 0.6425]

# Accuracy of TL trained models without finetuning for RBI 0.4, 0.5, 0.7 and 0.88 mean
Acc1 = [0.6905, 0.6660, 0.6410, 0.6390]

# Accuracy of TL trained models with finetuning for RBI 0.4, 0.5, 0.7 and 0.88 mean
Acc2 = [0.7835, 0.7315, 0.7080, 0.7025]
```

```python
x = [0.4, 0.5, 0.7, 0.88]

plt.figure()
plt.plot(x,Acc,'b', marker='o')
plt.plot(x,Acc1,'g', marker='v')
plt.plot(x,Acc2,'r', marker='^')
plt.grid()
plt.title('Comparison of accuracy of models for different RBI mean')
plt.xlabel('Mean of the RBI of all nodes in the training and testing set')
plt.ylabel('Accuracy')
plt.legend(['Accuracy on sepaarately trained models', 'Accuracy of TL model without FT', 'Accuracy of TL model with FT'], loc='upper right')
plt.show()
```

```python
#Time Computation

#Time computaion for separately trained models(training time + testing time for all samples)
T_sep = [1405.52, 1439.74, 1454.65, 1391.24]

#Time computation for TL without fine tuning (just testing on all samples)
T_TL_WFT = [0.385, 0.360, 0.326, 0.328]

#Time computation for TL with fine tuning
T_Tl_FT = [341.68, 318.34, 319.39, 319.50]
```
